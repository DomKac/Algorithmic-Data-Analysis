Używając danych z pliku `Auto.csv` z poprzedniej listy, utwórz i porównaj dwa modele regresji liniowej do przewidywania wartości `mpg`. W pierwszym modelu użyj `year` traktowanego jako zmienna **ciągła**. W drugim modelu użyj `year` traktowanego jako zmienna **kategoryczna**. Który model jest lepszy? Co jeśli byłoby znacznie więcej niż 13 wartości zmiennej year? Możesz zobaczyć przykłady w tym Jupyter notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from patsy import dmatrices, dmatrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the dataset
df = pd.read_csv('../lab2/Auto.csv')

# Convert all columns to numeric except 'name' and 'origin', coercing errors to NaN
df[df.columns.difference(['name', 'origin'])] = df[df.columns.difference(['name', 'origin'])].apply(pd.to_numeric, errors='coerce')
df['origin'] = df['origin'].astype('category')
```

```python
df.head()
```

```python
y_cont, X_cont = dmatrices('mpg ~ cylinders + displacement + horsepower + weight + acceleration + year + origin', data=df, return_type='dataframe')
y_cat, X_cat = dmatrices('mpg ~ cylinders + displacement + horsepower + weight + acceleration + C(year) + origin', data=df, return_type='dataframe')

# Model ze zmienną 'year' jako zmienną ciągłą
model_cont = sm.OLS(y_cont, X_cont)
results_cont = model_cont.fit()
print(results_cont.summary())

# Model ze zmienną 'year' jako zmienną kategoryczną
model_cat = sm.OLS(y_cat, X_cat)
results_cat = model_cat.fit()
print(results_cat.summary())

# MSE
y_pred_cont = results_cont.predict(X_cont)
y_pred_cat = results_cat.predict(X_cat)
mse_cont = mean_squared_error(y_cont, y_pred_cont)
mse_cat = mean_squared_error(y_cat, y_pred_cat)
print(f'MSE for continuous year: {mse_cont}')
print(f'MSE for categorical year: {mse_cat}')
```

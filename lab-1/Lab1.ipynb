```python
import string
import re

with open('The-Count-Of-Monte-Cristo.txt', 'r', encoding='utf-8') as file:
    text = file.read()
    text = text.lower() # Zamiana na małe litery
    text = text.replace("”", "").replace('“', "").replace('"', "").replace("'", "") # Usunięcie apostrofów i cudzysłowów
    text = text.replace("—", " ")
    text= text.replace("’", "").replace("‘", '')
    text = re.sub(r'\b\d+m\b', '', text) # Usunięcie liczb z m na końcu
    text_without_punctuation = text.translate(str.maketrans('', '', string.punctuation)) # Usunięcie znaków interpunkcyjnych
    words = text_without_punctuation.split() # Podzielenie tekstu na słowa
```

```python
print(words[:10])
print(len(words))
```

```python
# Usunięcie słów nieistotnych
filtered_words = [word for word in words if word not in stop_words]
print(len(filtered_words))
```

```python
# Stemming
from stemming.porter2 import stem
filtered_words = [stem(word) for word in filtered_words]
print(filtered_words[:100])
```

```python
# tworzenie słownika z unikalnych słów
unique_words = set(filtered_words)
words_dict = dict.fromkeys(unique_words, 0)
for word in filtered_words:
    words_dict[word] += 1

print(len(words_dict))
```

```python
# Posortowanie słownika według najczęściej występujących słów w porządku malejącym
words_dict_sorted = dict(sorted(words_dict.items(), key=lambda item: item[1], reverse=True))
print(words_dict_sorted)

# Usunięcie 5 najczęściej występujących słów
for key, value in list(words_dict_sorted.items())[:5]:
    print(f'{key}: {value}')
    words_dict_sorted.pop(key)
```

```python
import string
import re
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk

nltk.download('wordnet')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
print(stop_words)

print("rocks :", lemmatizer.lemmatize("rocks"))
print("corpora :", lemmatizer.lemmatize("corpora"))
 
# a denotes adjective in "pos"
print("better :", lemmatizer.lemmatize("better", pos ="a"))
```

```python
def get_words(text: str, do_lemmatization: bool = True, del_stop_words: bool = True) -> list:
    
    text = text.lower() # Zamiana na małe litery
    text = text.replace("”", "").replace('“', "")
    text= text.replace("’", "").replace("‘", '')
    text = text.replace('"', "").replace("'", "") # Usunięcie apostrofów i cudzysłowów
    text = text.replace("—", " ")
    text = re.sub(r'\b\d+m\b', '', text) # Usunięcie liczb z m na końcu
    text_without_punctuation = text.translate(str.maketrans('', '', string.punctuation)) # Usunięcie znaków interpunkcyjnych
    words = text_without_punctuation.split() # Podzielenie tekstu na słowa
    if del_stop_words:
        words = [word for word in words if word not in stop_words] # Usunięcie słów nieistotnych
    if do_lemmatization:
        words = [lemmatizer.lemmatize(word) for word in words]  # Lematyzacja

    return words
```

```python
with open('The-Count-Of-Monte-Cristo.txt', 'r', encoding='utf-8') as file:
    
    chapters = set()
    book_start_line = 0
    current_chapter = ""
    unique_words = set()

    while True:
        line = file.readline()
        if not line:
            break

        if line.strip().startswith('Chapter'):
            chapter_name = line.lstrip().rstrip()
            if chapter_name in chapters:
                current_chapter = chapter_name
                break
            else:
                chapters.add(chapter_name)


    chapters_dict = dict.fromkeys(chapters)
    print(chapters_dict)
    print(len(chapters_dict))
    print(current_chapter)
    
    while True:
        line = file.readline()
        if not line:
            break
        chapter_lines = ""

        while not line.strip().startswith('Chapter'):
            chapter_lines += line
            line = file.readline()
            if not line:
                break

        chapter_words = get_words(chapter_lines)
        unique_words.update(chapter_words)
        # chapters_dict[current_chapter] = chapter_words
        chapter_unique_words = set(chapter_words)
        chapter_words_occurrences = dict.fromkeys(chapter_unique_words, 0)
        for word in chapter_words:
            chapter_words_occurrences[word] += 1
        chapters_dict[current_chapter] = chapter_words_occurrences
        current_chapter = line.lstrip().rstrip()
```

```python
import math

def compute_tf_idf(chapters_dict):
    # Liczba rozdziałów (dokumentów)
    chapters_num = len(chapters_dict)
    
    # Słownik do przechowywania df (ile rozdziałów zawiera dane słowo)
    df = dict.fromkeys(unique_words, 0)

    # Obliczanie df (ile rozdziałów zawiera dane słowo)
    for word in unique_words:
        for chapter in chapters_dict:
            if word in chapters_dict[chapter]:
                df[word] += 1
    

    # Słownik do przechowywania tf-idf
    tf_idf_dict = {}
    
    # Obliczanie tf-idf
    for chapter, word_counts in chapters_dict.items():
        tf_idf_dict[chapter] = {}
        
        for word, count in word_counts.items():
            # Oblicz tf (term frequency)
            tf = chapters_dict[chapter][word]

            # Oblicz idf (inverse document frequency)
            idf = math.log(chapters_num / (1.0 + df[word]))
            # idf = 1
            # Oblicz tf-idf
            tf_idf = tf * idf
            tf_idf_dict[chapter][word] = tf_idf

    print(df['cristo']) 
    
    return tf_idf_dict

# Przykład użycia
# tf_idf_scores = compute_tf_idf(chapters_dict)
# print(tf_idf_scores)

tf_id_scores = compute_tf_idf(chapters_dict)
print(tf_id_scores)

```

stopu). Użyj tego podsumowania, aby wygenerować losowy akapit, który przypomina akapit z twojej książki.

```python
# Przygotowanie słownika do przechowywania następnych słów
with open('The-Count-Of-Monte-Cristo.txt', 'r', encoding='utf-8') as file:
    # deleted stop word and lemmatization
    words = get_words(file.read())
    # words = get_words(file.read(), do_lemmatization=False, del_stop_words=False)

following_words = dict.fromkeys(set(words), [])

for i, word in enumerate(words[:-1]):
    if following_words[word] == []:
        following_words[word] = [words[i+1]]
    else:
        following_words[word].append(words[i+1])

print(following_words['cristo'][:10])
```

```python
from collections import Counter

top5_following_words = {word: [w for w, _ in Counter(following_words[word]).most_common(5)]
                       for word in following_words}

print(top5_following_words['cristo'])
```

```python
import random

def generate_random_paragraph(start_word, max_length=50):
    paragraph = start_word
    current_word = start_word

    for _ in range(max_length - 1):
        next_words = top5_following_words.get(current_word, [])
        if not next_words:
            break
        # Wybór losowy spośród najczęstszych słów następujących
        current_word = random.choice(next_words)
        # current_word = next_words[0]
        paragraph += ' ' + current_word

    return paragraph

# Przykład użycia
start_word = 'cristo'  # Wybierz słowo początkowe
random_paragraph = generate_random_paragraph(start_word)
print(random_paragraph)
```
